```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv('loan_data.csv')
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
new_data=dataset.dropna(how='any')
```

```python
new_data.isna().sum()
```

```python
new_data['Dependents'].value_counts()
```

```python
new_data = new_data.replace(to_replace='3+', value=4)
```

```python
new_data['Dependents'].value_counts()
```

```python
new_data['Property_Area'].value_counts()
```

```python
new_data.replace({'Loan_Status':{'N':0,'Y':1},'Married':{'Yes':1,'No':0},'Self_Employed':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Education':{'Graduate':1,'Not Graduate':0},'Self_employed':{'No':0,'Yes':1},'Property_Area':{'Rural':0,'Urban':1,'Semiurban':2},'Loan_Status':{'N':0,'Y':1}},inplace=True)
```

```python
new_data.head()
```

```python
X = new_data.drop(['Loan_Status','Loan_ID'], axis=1)
Y = new_data['Loan_Status']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_predict=model.predict(X_train)
```

```python
X_train_acc=accuracy_score(X_train_predict,Y_train)
```

```python
print(X_train_acc)
```

```python
X_test_predict=model.predict(X_test)
X_test_acc=accuracy_score(X_test_predict,Y_test)
print(X_test_acc)
```
