```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_dataset = pd.read_csv('/content/mail_data.csv')
```

```python
mail_dataset.isnull().sum()
```

```python
#replace the null values with null string
mail_data = mail_dataset.where((pd.notnull(mail_dataset)),'')
```

```python
mail_data.shape
```

```python
mail_data.describe()
```

label encoding

```python
#label spam as 0,ham as 0
mail_data.loc[mail_data['Category'] == 'spam','Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham','Category',] = 1
```

0-->spam 1-->ham

```python
#seperating the data as texts and label
X =mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data to train and test
#random state is used to split in required way test_size is used to test amount of data 0.2 means 20% of data will be tested remaining will be trained
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
#transform the text data to feature that can be used as input to the Logistic Regression
feature_extraction  = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
# conver Y_train,Y_test values as integer
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

Training the model

Logistic regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

Evaluating the training model

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train,prediction_on_training_data)
```

```python
print("accuracy of training data",accuracy_on_training_data)
```

```python
prediction_on_testing_data = model.predict(X_test_features)
accuracy_on_testing_data = accuracy_score(Y_test,prediction_on_testing_data)
```

```python
print("accuracy of testing data",accuracy_on_testing_data)
```
